```python
import pandas as pd
import sqlite3 as sql
```

```python
database = "chinook.db"
```

```python
conn = sql.connect(database)
```

```python
pd.read_sql_query("select * from sqlite_master",conn)
```

```python
def query(q):
    return pd.read_sql_query(q,conn)
```

```python
q = """ select * from sqlite_master where type = "table"  """
query(q)
```

```python
q = "select * from album "
query(q)
```

```python
q = "select * from artist limit 5 "
query(q)
```

```python
q = "select * from customer limit 5 "
query(q)
```

```python
q = "select * from employee limit 5 "
query(q)
```

```python
q = "select * from album a inner join artist ar on  a.artist_id = ar.artist_id  "
query(q) 
```
